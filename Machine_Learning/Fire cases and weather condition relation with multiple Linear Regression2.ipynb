```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
```

```python
fire_data = pd.read_csv('drive/MyDrive/python/fire_data.csv', encoding='cp949')
weather = pd.read_csv('drive/MyDrive/python/weather.csv', encoding='cp949')
```

```python
raw_date = fire_data['date']

date = [i[:10] for i in raw_date]

cases = {}

for i in date:
    if i in cases:
        cases[i] += 1
    else :
        cases[i] = 1

sorted_cases =  sorted(cases.items(), reverse=True)

y_data = [i[1] for i in sorted_cases]
y_data = np.array(y_data).reshape((365, 1))


raw_weather = weather[['temp', 'high','low', 'hum', 'low hum', 'rain', 'high rain', 'wind', 'high wind']]
x_data = raw_weather
```

```python
print(sorted_cases)
print(raw_weather)
```

```python
data = {
    'date' : sorted_cases
}

num = [ i[1] for i in sorted_cases]

df3 = pd.DataFrame(num)

pd.concat([raw_weather, df3], axis = 1)
```

```python
df3.to_csv('test.csv',index=False)
```

```python
mlr = LinearRegression()
mlr.fit(x_data, y_data) 
```

```python
ans = mlr.predict(x_data)


accuracy = ans - y_data

print(accuracy)
```

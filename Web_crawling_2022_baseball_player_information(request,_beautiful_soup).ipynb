```python
url = ['https://elwlsek.tistory.com/2298?category=489300', 
       'https://elwlsek.tistory.com/2297?category=489300', 
       'https://elwlsek.tistory.com/2296?category=489300', 
       'https://elwlsek.tistory.com/2295?category=489300', 
       'https://elwlsek.tistory.com/2293?category=489300', 
       'https://elwlsek.tistory.com/2291?category=489300',
       'https://elwlsek.tistory.com/2286?category=489300',
       'https://elwlsek.tistory.com/2284?category=489300',
       'https://elwlsek.tistory.com/2276?category=489300',
       'https://elwlsek.tistory.com/2292' ]
#Hanwha, Kia, Lotte, NC, SSG, Kiwoom, SL, Doosan, KT, LG

team = ['한화', 'KIA', '롯데', 'NC', 'SSG', '키움', 'SL', '두산', 'KT', 'LG']
roaster = []
html = []
ans = []
```

```python
ans = []

for i in range(10) :
  html.append(requests.get(url[i]))

for i in range(10) :
  soup = bs(html[i].text,'html.parser')
  data1 = soup.find('div',{'class':'tt_article_useless_p_margin contents_style'})
  data2 = data1.findAll('td')
  cnt = 0
  block = 0

  for j in data2 :
    if block == 2 :
      break

    if "등번호" in j.text :
      block += 1
    
    cnt += 1
    
  print(cnt)

  pprint(data2[cnt - cnt % 7] )

  res = []
  point = 0

  for j in range(0, cnt - ( cnt % 7) ) :
    if '육성선수' in data2[j].text :
      res.append(team[i])
      continue

    if data2[j].text == '김동혁' :
      res.append('번호 없음')

    res.append(data2[j].text)

    if point == 1 and j % 8 == 7 :
      res.append(team[i])

    if '비고' in data2[j].text :
      point = 1

    if point == 0 and j % 7 == 6 :
      res.append(team[i])

    
  try :
    res.remove('비고')
  except :
    print('No 비고')
  
  try :
    res.remove('복귀')
  except :
    print('No 복귀')

  try :
    while '\u3000' in res :
      res.remove('\u3000')
  except :
    print('No \\u3000')

  np_res = np.array(res)
  
  print(len(np_res))
  '''
  try :
    np_res = np_res.reshape( int( len(np_res) / 8) , 8) 

  except :
    np_res = np_res.reshape( int ( len(np_res ) / 6 ) , 6)
  '''
  print(np_res)
  np_res = np_res.reshape( int( len(np_res) / 8) , 8) 
  ans.append(np_res)

#df.to_csv("2022_Hanwha_baseball_player_information.csv")  #saving as csv file
```

```python
print(ans)
```

```python
print(ans[1])
```

```python
final = []

for array in ans :
  for line in array :
    final.append(line)

print(final[0])

df = pd.DataFrame(final[1:], columns = ['등번호', '선수명', '포지션', '연봉', '생년월일', '체격', '출신교', '팀명'] )

df = df.set_index('팀명')
df.to_csv("2022_baseball_player_information.csv", encoding = 'cp949')  #saving as csv file
```

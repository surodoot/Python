```python
import matplotlib.pyplot as plt

years = list(range(1950, 2011, 10))
gdp = [ 67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.plot( years, gdp, color = 'green', marker = 'o', linestyle = 'solid')

plt.title('GDP per capita')

plt.ylabel("dollars")
plt.savefig("gdp_per_capita.png", dpi = 600 )
plt.show()
```

```python
plt.plot( gdp, years, color = 'red', marker = 'o', linestyle = 'solid')
plt.show()
```

```python
x = [ i for i in range(-10, 10)]  # same as "x = list(range(-10,10))"
y = [ 2 * i for i in x ] 

plt.plot(x, y, marker='o', color = 'cyan')

plt.axis([ -20, 20, -20, 20])
plt.show()
```

```python
x = list( range( -20, 20))

y1 = [ i * 2 for i in x ]

y2 = [ i ** 2 for i in x ]

y3 = [-i ** 2 - 5 for i in x ]

plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(list(range(-20, 20)))

y1 = x * 2
y2 = x ** 2
y3 = -x ** 2 - 5

plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30])
plt.show()
```

```python
domain = np.arange(20)

x = domain
y = domain ** 2
z = domain ** 3

plt.plot ( x, x, label = 'Linear ' )
plt.plot ( x, y, label = 'Quadratic')
plt.plot ( x, z, label = 'Cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('My plot')
plt.legend()
plt.show()
```

```python
import math as m
import matplotlib.pyplot as plt

x = []
y = []

for angle in range(360) :
  x.append(angle)
  y.append( m.sin(m.radians(angle)))

plt.plot(x, y)
plt.title("Sine Wave")
plt.show()
```

```python
import math as m
import matplotlib.pyplot as plt

x = range(360)
y = [ m.sin(m.radians(i)) for i in x]

plt.plot(x, y)
plt.title("Sine Wave")
plt.show()
```

```python
years = list( range(1950, 2011, 10) )
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.bar( range(len(years)), gdp)
plt.title("GDP per capita")
plt.ylabel("dollars")

plt.xticks(range(len(years)), years)
plt.show()
```

```python
years = list( range(1965, 2016, 10))
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [890, 5120, 11500, 42130, 40560, 38780]
ch = [100, 200, 290, 540, 1760, 7940]

x_range = ( range(len(years)))

plt.bar(x_range, jp, width = 0.25, color = 'red')
plt.bar(x_range, ko, width = 0.25, color = 'blue')
plt.bar(x_range, ch, width = 0.25, color = 'orange')

plt.xticks( range(len(years)), years)
plt.show()
```

```python
xData = np.arange(20, 50, 0.25)
yData = xData + 5 * np.random.randn(120)

plt.scatter(xData, yData)
plt.title('Real Age vs Physical Age')
plt.xlabel('Real Age')
plt.ylabel('Physical Age')

plt.savefig("Real_Age_vs_Physica_Age.png", dpi = 600)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
line_fitter = LinearRegression()
line_fitter.fit(xData.reshape(-1, 1), yData)

plt.scatter(xData, yData)
plt.plot(xData,line_fitter.predict(xData.reshape(-1,1)), color = 'red') #Linear regression
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
x_min=np.array(weight).min()
x_max=np.array(weight).max()
print(x_min,x_max)
plt.hist(weight)
plt.show()

print(max(weight), min(weight))

plt.hist(weight, label = 'bins = 10')
plt.hist(weight, bins = 30, label = 'bins = 30')
plt.legend()
plt.show()

plt.hist(weight, label = 'bins = 10', alpha = 0.3)
plt.hist(weight, bins = 15, label = 'bins = 15', color = 'red', alpha = 0.6)
plt.hist(weight, bins = 30, label = 'bins = 30', color = 'yellow', alpha = 0.7)
plt.legend()
plt.show()
```

```python
url = 'https://elwlsek.tistory.com/2298?category=489300'

from bs4 import BeautifulSoup as bs 
from pprint import pprint
import requests
html = requests.get(url)
pprint(html.text)
```

```python
soup = bs(html.text,'html.parser')
pprint(soup)
```

```python
data1 = soup.find('div',{'class':'tt_article_useless_p_margin contents_style'})
pprint(data1)
#data2 = data1.
```

```python
data2 = data1.findAll('td')
#size = int ( len(data2) / 2 )
#pprint(data2[:size])
pprint(data2)
```

```python
cnt = 0
block = 0

for i in data2 :

  if "청주우암초-청주중-청주고-한화-경찰" in i.text :
    break
  
  cnt += 1
  
print(cnt)
```

```python
pprint(data2[cnt])
res = []

for i in range(0, cnt + 1) :
  if not("육성선수" in data2[i].text ) :
    res.append(data2[i].text)

res.remove('비고')
res.remove('복귀')
while '\u3000' in res :
  res.remove('\u3000')


print(res)
```

```python
import numpy as np

np_res = np.array(res)
np_res = np_res.reshape( 84, 7)
```

```python
import pandas as pd


df = pd.DataFrame(np_res[1:], columns = np_res[0])

df.set_index('등번호')
```

```python
df.to_csv("2022_Hanwha_baseball_player_information.csv")  #saving as csv file
```
